```python
import pandas as pd

file_age = 'https://raw.githubusercontent.com/LondonEnergyMap/rawdata/master/building_stock/dwelling-period-built-2015-lsoa-msoa.csv'
file_type = 'https://raw.githubusercontent.com/LondonEnergyMap/rawdata/master/building_stock/dwelling-property-type-2015-lsoa-msoa.csv'
file_wallgas = 'https://raw.githubusercontent.com/LondonEnergyMap/rawdata/master/building_stock/solidwall_offgas_ldn.csv'
file_pcodelsoa = 'https://raw.githubusercontent.com/LondonEnergyMap/rawdata/master/building_stock/postcode_lsoa_ldn.csv'

dfage = pd.read_csv(file_age)
dftype = pd.read_csv(file_type)
dfwallgas = pd.read_csv(file_wallgas)
dfpcodelsoa = pd.read_csv(file_pcodelsoa)
```

```python
df = pd.melt(dftype_lsoa, id_vars =typecols[0:3], value_vars =typecols[4:36], var_name ='prop_type', value_name='ndwellings')
```

```python
df = df[(~df['BAND'].str.contains('All'))]
df.head()
```

```python
droprows = ['BUNGALOW',
            'FLAT_MAIS',
            'HOUSE_TERRACED',
            'HOUSE_SEMI',
            'HOUSE_DETACHED',
            'ALL_PROPERTIES']
df = df[(~df.prop_type.isin(droprows))]
df.prop_type.unique()
```

```python
# extract number from column

df['nbeds'] = df.prop_type.str.extract('(\d+)')
```

```python
df.head()
```

```python
df['prop_type'] = df.prop_type.str.lower()
```

```python
types = ['bungalow', 'house', 'flat']
forms = ['terraced', 'semi', 'detached']

df['prop'] = df.prop_type
df['form'] = df.prop_type

import numpy as np
for i in range(3):
    df['prop']= np.where(df.prop_type.str.contains(types[i]), types[i], df.prop)
    df['form']= np.where(df.prop_type.str.contains(forms[i]), forms[i], df.form)
```

```python
df.drop(columns=['prop_type'], inplace=True)
df.head()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.columns
```

```python
df['la'] = df.area_name.str.split().str[0]
```

```python
df['ndwellings'] = pd.to_numeric(df.ndwellings, errors='coerce')
temp = df[['la', 'prop', 'ndwellings']]
temp.groupby(by=['la', 'prop']).sum()
```
