```python
import pandas as pd

filepath = 'http://raw.githubusercontent.com/LondonEnergyMap/rawdata/master/consumption_postcode/gas/Postcode_level_gas_estimates_2015_experimental.csv'

df_all = pd.read_csv(filepath)
df_all.head()
df_all.shape
```

```python
# select london only postcodes
df_all = df_all.drop_duplicates()
df1 = df_all.copy()
df1['firstletter'] = df_all['POSTCODE'].str[0]
df1 = df1[(df1.firstletter.str.startswith('N')) |
           (df1.firstletter.str.startswith('E')) |
           (df1.firstletter.str.startswith('S')) |
           (df1.firstletter.str.match('W'))]
df1.drop(columns='firstletter', inplace=True)          
df1.head()
```

```python
# select inbetween poscodes

df2 = df_all[(df_all['POSTCODE'].str[0:2].str.startswith('NW')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('SE')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('SW')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('EC')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('WC')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('WD')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('EN')) |            
           (df_all['POSTCODE'].str[0:2].str.startswith('HA')) |            
           (df_all['POSTCODE'].str[0:2].str.startswith('IG')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('UB')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('TW')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('KT')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('SM')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('CR')) |
           (df_all['POSTCODE'].str[0:2].str.startswith('BR')) |             
           (df_all['POSTCODE'].str[0:2].str.startswith('DA')) |             
           (df_all['POSTCODE'].str[0:2].str.startswith('RM')) ]
```

```python
df = df1.append(df2, ignore_index=True)
df.shape
```

```python
df.columns=['pcode', 'gas', 'nmeter', 'avg', 'mid']
```

```python
dfm = df.merge(df_epc, on=['pcode', 'pcode'])
dfm_temp = dfm[dfm.pcode.duplicated(keep=False)]
pd.set_option('display.max_columns', 100)
dfm_temp.sort_values(by=['pcode'])
```

```python
dfm.drop_duplicates(inplace=True)
dfm.shape
```

```python
# define this dataframe as exact matches
dfm['pcode_match']='exact'
```

```python
# check which ones not matched

# dif = df[(~df.pcode.isin(dfm.pcode)) & (~df_epc.pcode.isin(dfm.pcode))]
remain = df_epc.copy()
remain = remain[(~remain.pcode.isin(dfm.pcode))]
remain.shape
```

```python
remain.head()
```

```python
# split remaining unmatched postcodes and match only first half
remain['pcode_match'] = 'partial'
pc = remain.pcode.str.split(' ', n = 1, expand = True)
remain['pc1'] = pc[0]
remain['pc2'] = pc[1]
remain.head()
```

```python
# same for postcode df by dropping entries with more than 3 letters
dfshort = df[df.pcode.str.len()<=3]
#dfshort = dfshort.rename(columns={'pcode':'pcshort'})
```

```python
# the remaining epc certificates with first part of pcodes
temp = dfshort.merge(remain, left_on='pcode', right_on='pc1')
temp.shape
```

```python
temp.drop_duplicates(inplace=True)

temp.shape
```

```python
temp.head()
```

```python
# swap pcode_x and pcode_y columns then drop pcode_x
p = temp.pop('pcode_y')
temp['pcode_x']=p
```

```python
temp = temp.rename(columns={'pcode_x':'pcode'})
```
