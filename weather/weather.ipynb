```python
import pandas as pd

filepath = 'https://raw.githubusercontent.com/LondonEnergyMap/rawdata/master/weather/heathrow_weather.csv'

df = pd.read_csv(filepath)
df.head()
```

```python
# keep only 2000 onwards data

df = df[(df.yyyy>=2000) & (df.yyyy<=2018)]
df.head(12)
```

```python
df['tmid'] = (df.tmax - df.tmin)/2 + df.tmin
```

```python
temp = df[df.yyyy == 2000]

import seaborn as sns
from matplotlib import pyplot as plt

sns.kdeplot(temp.tmax)
sns.kdeplot(temp.tmin)
sns.kdeplot(temp.tmid)
plt.show()
```

```python
for i in range(2000, 2018):
    temp = df[df.yyyy == i]
    sns.kdeplot(temp.tmid)
    
plt.show()
    
```

```python
# how to decide if hot or cold winter? use max and min temperature, and whether long or short spells?

sns.boxplot(df.yyyy, df.tmid)
plt.show()
```
